This file views the airport dataset used in the project.

```python
from __future__ import annotations

import pandas as pd
```

```python
# read file as dataset

file_location = "airports.parquet"
flight_dataframe = pd.read_parquet(file_location)
print(flight_dataframe.columns)

print(flight_dataframe.head())

print("number of rows:", len(flight_dataframe))
```

```python
# how many in uk of each type
uk_airports = flight_dataframe[flight_dataframe["iso_country"] == "GB"]
print(uk_airports)
```

```python
# print regions
print(uk_airports["iso_region"].value_counts())
```
